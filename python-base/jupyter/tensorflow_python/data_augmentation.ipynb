```python
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from keras.preprocessing import image
import keras.backend as K
import os
import glob
import numpy as np
#  AI训练中为了增强模型训练，通过各种方法增强数据，
# 图像数据增强一般有：平移、反转、旋转、放大缩小、截取等。
# python 3.7 keras 2.3.1
```

```python
def print_result(path):
#     展示图片
    name_list=glob.glob(path)
    fig=plt.figure(figsize=(12,16))
    for i in range(3):
        img=Image.open(name_list[i])
        sub_img=fig.add_subplot(131+i)
        sub_img.imshow(img)
img_path='D:\\test\\in\\target\\*'
in_path='D:\\test\\in'
out_path='D:\\test\\output'
name_list=glob.glob(img_path)
print(name_list)
print_result(img_path)
```

```python
# 图片缩小 指定 target_size后，所有图像都变为相同大小
datagen=image.ImageDataGenerator()
gen_data=datagen.flow_from_directory(in_path,batch_size=1,shuffle=False,
                                     save_to_dir=out_path,save_prefix='gen',target_size=(500,500))
for i in range(len(gen_data)):
    gen_data.next()
```

```python
print_result(out_path+"\\*")
```

```python
# 图片角度变换
datagen=image.ImageDataGenerator(rotation_range=60)
gen=image.ImageDataGenerator()
data=gen.flow_from_directory(in_path,batch_size=1,class_mode=None,shuffle=True,target_size=(500,500))
np_data=np.concatenate([data.next() for i in range(data.n)])
datagen.fit(np_data)
gen_data=datagen.flow_from_directory(in_path,batch_size=1,shuffle=False,save_to_dir=out_path,
                                     save_prefix='xz',target_size=(500,500))
for i in range(3):
    gen_data.next()
print_result(out_path+'\\*')
```

```python
# 图片平移变换
datagen=image.ImageDataGenerator(width_shift_range=0.3,height_shift_range=0.3)
gen=image.ImageDataGenerator()
data=gen.flow_from_directory(in_path,batch_size=1,class_mode=None,shuffle=True,target_size=(600,600))
np_data=np.concatenate([data.next() for i in range(data.n)])
datagen.fit(np_data)
gen_data=datagen.flow_from_directory(in_path,batch_size=1,shuffle=False,save_to_dir=out_path,
                                     save_prefix='py',target_size=(600,600))
for i in range(3):
      gen_data.next()
print_result(out_path+'\\*')
```
